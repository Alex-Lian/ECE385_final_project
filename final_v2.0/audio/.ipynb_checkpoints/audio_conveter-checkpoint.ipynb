```python
#record the wav file
depth = data.shape[0] + 1
print(depth)

#outFile = open("mono_"+filename+"_hex_rom.mif", 'w')
outFile = open("happy_background.mif", 'w')
outFile.write("WIDTH = 16;\n")
outFile.write("DEPTH = %d;\n" %depth)
outFile.write("ADDRESS_RADIX = HEX;\n")
outFile.write("DATA_RADIX = HEX;\n")
outFile.write("CONTENT BEGIN\n")
outFile.write("\n")


for i in range(data.shape[0]):
    outFile.write("%X: %04X;\n" %(i+1, ( int(round(data[i,1])) & 0xffff )))

outFile.write("END;\n")
outFile.close()
```
